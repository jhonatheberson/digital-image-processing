```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt 
```

```python
img  = np.array([[0, 5, 7, 7, 5, 8, 7, 8],[7, 2 ,6 ,2, 6, 5, 6, 8], [6, 9, 7, 7, 0, 7, 2, 7], [6, 6, 1 ,7 ,6 ,7 ,7, 5], [9, 6, 0 ,7 ,8 ,2 ,6 ,7] ,[2, 8 ,8 ,2 ,7 ,6 ,7 ,8], [7, 3, 2, 6 ,1 ,7 ,5 ,8], [9, 9, 5 ,6 ,7 ,7 ,7 ,7]], dtype = np.int8)
print(img.shape)
print(type(img))

plt.plot(),plt.imshow(img, 'gray'),plt.title('ORIGINAL')
# plt.savefig('Figura_5.png')
cv.imwrite('Figura_5.png', img)
plt.show()
```

```python
img_2 = cv.imread("./Figura_5.png",0)
print(type(img_2))

histr = cv.calcHist([img_2],[0],None,[8],[0,255])
img_equ = cv.equalizeHist(img_2)
plt.plot(),plt.imshow(img_equ, 'gray'),plt.title('Equalized')
# plt.savefig('Figura_5-b.png')
cv.imwrite('Figura_5-b.png', img_equ)
plt.show()

equ = cv.calcHist([img_equ],[0],None,[8],[0,255])
plt.plot(), plt.plot(histr), plt.title('a) - histogram')
plt.savefig('5-a.png')
plt.show()

plt.plot(), plt.plot(equ), plt.title('b) - equalized histogram')
plt.savefig('5-b.png')
plt.show()
```
