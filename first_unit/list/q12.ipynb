```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt 
```

```python
img = cv.imread("./Figura_12.png",0)

plt.plot(),plt.imshow(img, 'gray'),plt.title('ORIGINAL')
# plt.savefig('Figura_5.png')
# cv.imwrite('Figura_12.png', img)
plt.show()
```

```python
img_2 = cv.imread("./Figura_12.png",0)
# print(type(img_2))

histr = cv.calcHist([img_2],[0],None,[255],[0,255])
img_equ = cv.equalizeHist(img_2)
plt.plot(),plt.imshow(img_equ, 'gray'),plt.title('Equalized')
# plt.savefig('Figura_5-b.png')
cv.imwrite('Figura_12-b.png', img_equ)
plt.show()

equ = cv.calcHist([img_equ],[0],None,[255],[0,255])
plt.plot(), plt.plot(histr), plt.title('a) - histogram')
plt.savefig('12-a.png')
plt.show()

plt.plot(), plt.plot(equ), plt.title('b) - equalized histogram')
plt.savefig('12-c.png')
plt.show()
```
