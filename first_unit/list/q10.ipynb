```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt 
```

```python
img = cv.imread("./Figura_5.png",0)
img_low = img.copy()
img_hight = img.copy()
for i in range(8):
    for j in range(8): 
        img_low[i][j] = img[i][i] & 11111110 #zerando menos signifcativo
        print(bin(img[i][i]))
for i in range(8):
    for j in range(8): 
        img_hight[i][j] = img[i][i] & 0b01111111 #zerando menos signifcativo
hist = cv.calcHist([img],[0],None,[8],[0,255])
# equ = cv.calcHist([img_equ],[0],None,[8],[0,255])
plt.plot(), plt.plot(hist), plt.title('10)')
plt.savefig('10.png')
plt.show()

hist_low = cv.calcHist([img_low],[0],None,[8],[0,255])
plt.plot(), plt.plot(hist_low), plt.title('10 - a)')
plt.savefig('10-a.png')
plt.show()

hist_hight = cv.calcHist([img_hight],[0],None,[8],[0,255])
plt.plot(), plt.plot(hist_hight), plt.title('10 - b)')
plt.savefig('10-b.png')
plt.show()
```
